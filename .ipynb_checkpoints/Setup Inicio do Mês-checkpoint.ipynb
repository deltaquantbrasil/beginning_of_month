```python
import pandas as pd
import numpy as np
import yfinance as yf

import pandas_ta as ta

import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')
```

# Obtendo os dados

```python
df = yf.download("BOVA11.SA", period='5y')
```

# Gerando os retornos diários

```python
data = df.copy()
data["return"] = data["Adj Close"].pct_change()
data.head()
```

# Criando variáveis dia do mês e ano

```python
data["day_of_month"] = data.index.day
data["year"] = data.index.year
```

# Agrupando por dia e obtendo a média dos retornos

```python
grouped_by_day = data.groupby("day_of_month")['return'].mean()
```

```python
grouped_by_day.plot.bar(title="Média de retornos por dia do mês", figsize=(12,6), color='tab:green', alpha=0.7);
```

* Podemos observar que os 6 primeiros dias do mês tem, em média, retornos positivos com diminuição gradativa até o quinto dia. Logo estaremos utilizando apenas os quatro primeiros dias para gerar os resultados.

# Gerando os retornos dos 4 primeiros dias do mês

```python
first_days = 4
data["first_days_returns"] = 0.0
data.loc[data.day_of_month <= first_days, "first_days_returns"] = data[data.day_of_month <= first_days]['return']
```

```python
(
    data.groupby("year")
    .first_days_returns.mean()
    .plot.bar(title="Média de retornos da estratégia por ano", figsize=(12,6), color='tab:blue', alpha=0.7)
);
```

* No gráfico acima, podemos ver que a média de retornos agrupados por ano tem retorno positivo ao longo dos anos.

# Verificando os retornos acumulados por ano

```python
(
    data.groupby("year")
    .first_days_returns.sum()
    .cumsum()
    .plot(title="Retornos Acumulados por Ano", figsize=(12,6), color='tab:red')
    .axhline(0, color='black', ls='--', alpha=0.5)
);
```

# Verificando os retornos acumulados por dia

```python
(
    data.first_days_returns.cumsum()
    .plot(title="Soma Acumulada de Retornos Diários",figsize=(12,6), color='tab:orange')
    .axhline(0, color='black', ls='--', alpha=0.5)
);
```

## <span style="color:red">Podemos concluir que o setup foi lucrativo ao longo dos anos.</span>
